```python
%matplotlib inline
%run convenience.ipynb
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

np.set_printoptions(precision=2,suppress=True) # numpy output options

pi=np.pi
j=1j
```

```python
#RC
#sys = signal.lti([1/(R*C)], [1,1/(R*C)])
#sys_string='RC'#identifier for graphs

C=1e-7
L=1e-2
R=50#400 or 50 ohm

#RLC
sys = signal.lti([1,0], [1,1])
sys_string='3a'#identifier for graphs

time_scale=5*L/R#how long graphs should go for; scales inputs accordingly
t=np.linspace(0,10*L/R,500)#set the number of points OR
#t=np.arange(0,5*L/R,time_scale/500)#set the step size
```

```python
#signal.lti expects a transfer function of the form (num,den), 
#where num and den are lists from highest order to lowest.
#Example: H=(2*s^2+3*s+4)/(s^3+5*s^2+6*s+7) is entered as
#sys = signal.lti([2,3,4],[1,5,6,7])
```

```python
_t,s=signal.step(sys,T=t)
timeplot(t,s,label=sys_string+' step response (V)')
```

```python
freqs=np.logspace(3,5,500)
_w, mag, phase = signal.bode(sys,w=2*pi*freqs)#Bode takes frequencies in rad/s and returns magnitudes in dB
logplot(freqs,10**(mag/20),phase,title=sys_string+' Bode plot',margins=0.4)
```

```python
pzmap(sys)
```

The system acts as a simple high pass filter. The function has one negative real pole, meaning that it decays exponentially.

```python
%matplotlib inline
%run convenience.ipynb
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

np.set_printoptions(precision=2,suppress=True) # numpy output options

pi=np.pi
j=1j
```

```python
sys = signal.lti([1], [1,-.01,1])
t=np.linspace(0,300,1000)
```

```python
_t,s=signal.step(sys,T=t)
timeplot(t,s,label='step response (V)')
```

```python
pzmap(sys)
```

```python
pzmap(sys,axis=[-.01,.01,-1.5,1.5])#rescaling
```

```python
Kp = -.5
psys = signal.lti([Kp,0], [1,-.01,(1+Kp)])
# pzmap(psys)
pzmap(psys,axis=[-.01,.01,-1.5,1.5])#rescaling
```

```python
_t,sp=signal.step(psys,T=t)
timeplot(t,sp,label='step response (V)')
```

```python
Ki = 0
isys = signal.lti([Ki,0], [1,-.01,1,Ki])
pzmap(isys)
# pzmap(isys,axis=[-.01,.01,-1.5,1.5])#rescaling
```

```python
Kd = 2
dsys = signal.lti([Kd,0],[1,(Kd-.01),1])
pzmap(dsys)
```

```python
_t,dp=signal.step(dsys,T=t)
timeplot(t,dp,label='step response (V)')
```

In this case, derivative control was most effective. This is because with derivative control, I was able to adjust the value of the real part of the poles, and make them negative. This allows our function to decay quickly.
